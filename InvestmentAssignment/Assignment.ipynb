## CheckPoint 2: Funding Type Analysis

### Average funding amount of venture type

```python
master_frame['raised_amount_usd'] = master_frame['raised_amount_usd'].astype(float)
venture = ( master_frame['funding_round_type']=='venture' )
Avg_venture = ((( master_frame.loc[venture,'raised_amount_usd'].mean())/1000000).round(4))
```

### Average funding amount of angel type

```python
angel = ( master_frame['funding_round_type'] == 'angel' )
Avg_angel = ((( master_frame.loc[angel,'raised_amount_usd'].mean())/1000000).round(4))
```

### Average funding amount of seed type

```python
seed = ( master_frame['funding_round_type'] == 'seed')
Avg_seed = ((( master_frame.loc[seed, 'raised_amount_usd'].mean())/1000000).round(4))
```

### Average funding amount of private equity type

```python
equity = ( master_frame['funding_round_type'] == 'private_equity')
Avg_equity = (((master_frame.loc[equity, 'raised_amount_usd'].mean())/1000000).round(4))
```

```python
print ('Average funding amount of Venture type = $ %sM' %Avg_venture)
print ('Average funding amount of Angel type = $ %sM' %Avg_angel)
print ('Average funding amount of Seed type = $ %sM' %Avg_seed)
print ('Average funding amount of Private Equity type = $ %sM' %Avg_equity)
```

```python
Avg_funding = {'Venture':Avg_venture,'Angel':Avg_angel,'Seed':Avg_seed,'Private_Equity':Avg_equity}
type_list = []
for key, value in Avg_funding.items():
    if value >= 5 and value <=15:
        type_list.append(key)
        
Investment_type = max(type_list)

print ('Considering that Spark Funds wants to invest between 5 to 15 million USD per investment round,\
 the most suitable  Investment type is =', Investment_type)
```

## Checkpoint 5: Sector Analysis 2
- **Creating the Dataframes D1, D2 and D3 with respect to top 3 countries.**

```python
# DataFrame D1 which will have the data corresponding to USA
filter1 = np.where( (top3_master_frame.country_code == 'USA')&(top3_master_frame.raised_amount_usd_million >= 5)&
                   (top3_master_frame.raised_amount_usd_million <= 15))
D1 = pd.DataFrame(top3_master_frame.iloc[filter1])

# Extracting the total amount invested in each main sector
D1_sectorwise_sum =pd.DataFrame(D1.groupby('main_sector').sum()['raised_amount_usd'])
D1_sectorwise_sum.rename(columns={'raised_amount_usd':'main_sectorwise_sum'}, inplace=True)
D1_sectorwise_sum.reset_index(inplace=True)

# Extracting the total number (or count) of investments for each main sector
D1_sectorwise_count = pd.DataFrame(D1.groupby('main_sector').count()['raised_amount_usd'])
D1_sectorwise_count.rename(columns={'raised_amount_usd':'main_sectorwise_count'}, inplace=True)
D1_sectorwise_count.reset_index(inplace=True)

D1 = pd.merge(D1,D1_sectorwise_count, how='inner', on='main_sector')
D1 = pd.merge(D1,D1_sectorwise_sum, how='inner', on='main_sector')

# Cross verifying the dataframe's by priting
print ('Filtered Country Code:',D1.country_code.unique())
print (D1.describe())
D1.head()
```

```python
# DataFrame D2 which will have the data corresponding to GBR
filter2 = np.where( (top3_master_frame.country_code == 'GBR')&(top3_master_frame.raised_amount_usd_million >= 5)&
                   (top3_master_frame.raised_amount_usd_million <= 15))
D2 = pd.DataFrame(top3_master_frame.iloc[filter2])

# Extracting the total amount invested in each main sector
D2_sectorwise_sum =pd.DataFrame(D2.groupby('main_sector').sum()['raised_amount_usd'])
D2_sectorwise_sum.rename(columns={'raised_amount_usd':'main_sectorwise_sum'}, inplace=True)
D2_sectorwise_sum.reset_index(inplace=True)

# Extracting the total number (or count) of investments for each main sector
D2_sectorwise_count = pd.DataFrame(D2.groupby('main_sector').count()['raised_amount_usd'])
D2_sectorwise_count.rename(columns={'raised_amount_usd':'main_sectorwise_count'}, inplace=True)
D2_sectorwise_count.reset_index(inplace=True)

D2 = pd.merge(D2,D2_sectorwise_count, how='inner', on='main_sector')
D2 = pd.merge(D2,D2_sectorwise_sum, how='inner', on='main_sector')

# Cross verifying the dataframe's by priting
print ('Filtered Country Code:',D2.country_code.unique())
print (D2.describe())
D2.head()
```

```python
# DataFrame D3 which will have the data corresponding to IND
filter3 = np.where( (top3_master_frame.country_code == 'IND')&(top3_master_frame.raised_amount_usd_million >= 5)&
                   (top3_master_frame.raised_amount_usd_million <= 15))
D3 = pd.DataFrame(top3_master_frame.iloc[filter3])

# Extracting the total amount invested in each main sector
D3_sectorwise_sum =pd.DataFrame(D3.groupby('main_sector').sum()['raised_amount_usd'])
D3_sectorwise_sum.rename(columns={'raised_amount_usd':'main_sectorwise_sum'}, inplace=True)
D3_sectorwise_sum.reset_index(inplace=True)

# Extracting the total number (or count) of investments for each main sector
D3_sectorwise_count = pd.DataFrame(D3.groupby('main_sector').count()['raised_amount_usd'])
D3_sectorwise_count.rename(columns={'raised_amount_usd':'main_sectorwise_count'}, inplace=True)
D3_sectorwise_count.reset_index(inplace=True)

D3 = pd.merge(D3,D3_sectorwise_count, how='inner', on='main_sector')
D3 = pd.merge(D3,D3_sectorwise_sum, how='inner', on='main_sector')

# Cross verifying the dataframe's by priting
print ('Filtered Country Code:',D3.country_code.unique())
print (D3.describe())
D3.head()
```

## Checkpoint 6: Plots

```python
# Grouping and sorting the cleaned master frame based on 'funding_round_type'
master_frame_grouped = master_frame_clean.groupby('funding_round_type')
master_frame_grouped = master_frame_grouped['raised_amount_usd'].sum().sort_values(ascending=False)

types = master_frame_grouped[['venture','angel','private_equity','seed']]
total_investments = master_frame_grouped[['venture','angel','private_equity','seed']].sum()

# Geting the fraction of total investments(globally) in angel, venture, seed, and private equity in pecentages
fractions = types.divide(total_investments)*100
fractions.round()
```

```python
# Creating the dataframe for the plot1
df_plot1 = pd.DataFrame(master_frame_grouped[['venture','angel','private_equity','seed']])
df_plot1['fractions']=fractions.round()

df_plot1
```

**1. A plot showing the fraction of total investments (globally) in angel, venture, seed, and private equity, and the average amount of investment in each funding type.**

```python
labels = df_plot1.index
colors = ['coral','blue', 'cyan', 'gold']
plt.figure(figsize=(12,14))
plt.pie(df_plot1['fractions'], labels= labels, colors=colors, startangle=90,
        autopct='%.1f%%',wedgeprops={ 'linewidth' : 4, 'edgecolor' :'white'})
plt.show()
```

```python
# Getting the funding types and also the average values and ploting the bar graph
master_frame_types = master_frame_copy[master_frame_copy.funding_round_type.isin(['venture', 'seed', 'angel', 'private_equity'])]
funding_types = master_frame_types.funding_round_type.value_counts()

plt.figure(figsize=(20, 16))
ax1 = plt.subplot(4, 4, 1)
ax2 = plt.subplot(4, 4, 3)
ax1.set(xlabel='Funding Types',ylabel='Average Funding in Million',title='Average investments based on Funding Types')
ax2.set(xlabel='Funding Types',ylabel='Average Funding in Million',title='Average investments based on Funding Types')
sns.barplot(x=funding_types.index, y=[Avg_venture, Avg_seed, Avg_angel, Avg_equity], ax=ax1)
sns.barplot(x=funding_types.index[1:3], y=[Avg_seed, Avg_angel], ax=ax2)

plt.show()

```
