#### Column To Drop

* We have following columns to drop as they has only one value in them, which is not suitable for analysis.
    - tax_liens
    - delinq_amnt
    - chargeoff_within_12_mths
    - acc_now_delinq
    - application_type
    - policy_code
    - collections_12_mths_ex_med
    - initial_list_status
    - pymnt_plan
    
* Apart form these columns, we should drop following columns:
    - **id** - It is unique for the table, and could be used as a primary key for the table. But, as we don't have to add any  other data with the existing table, therefore we should remove it.
    - **member_id** - Same as the above.
    - **url** - It should be dropped, as the only extractable information from this column is applicant's id, which is already                 been given.
    - **desc** - It should be dropped, as whatever could be extracted from here, is already in 'purpose' column.

Now, there are three types of columns in the dataframe, namely:

- Customer info
- Customer behaviour
- Loan characterstics

Loan characterstics can be applied before and during the loan award. But Customer behaviour can be applied only after the award of loan. Therefore, to study driving factors to default, we should exclude customer behaviours label.

```python
columns_to_dropped = ['id','member_id','tax_liens','delinq_amnt','chargeoff_within_12_mths','acc_now_delinq','application_type',
                      'policy_code', 'collections_12_mths_ex_med', 'initial_list_status', 'pymnt_plan', 'url', 'desc']
loan.drop(columns_to_dropped, axis=1, inplace=True)
```

```python
# Creating a list of customer_behaviour labels
cust_behaviour = ['acc_now_delinq', 'chargeoff_within_12_mths', 'collections_12_mths_ex_med', 'delinq_2yrs', 'delinq_amnt',
                 'inq_last_6mths', 'last_pymnt_amnt', 'mths_since_last_delinq', 'out_prncp', 'out_prncp_inv', 'recoveries',
                 'revol_bal', 'revol_util', 'total_rec_int', 'total_rec_late_fee', 'total_rec_prncp']

# As some columns in customer behaviour list has already dropped, therefore excluding them
cust_behaviour = list(set(cust_behaviour) - set(columns_to_dropped))
```

```python
# Dropping customer behaviour columns
loan.drop(cust_behaviour, axis=1, inplace=True)
```

As we have to determine driving factors, of default. Therefore, we should exclude loans with status **current**.

```python
indices_to_drop = loan[loan.loan_status=='Current'].index
loan.drop(indices_to_drop, axis=0, inplace=True)
```
